```python
import tushare as ts
a = 'ts.get_report_data'
"""
eval

　　功能：将字符串str当成有效的表达式来求值并返回计算结果。

　　语法： eval(source[, globals[, locals]]) -> value

　　参数：

　　　　source：一个Python表达式或函数compile()返回的代码对象

　　　　globals：可选。必须是dictionary

　　　　locals：可选。任意map对象
"""
list = ["ts.forecast_data","ts.fund_holdings"]
tab_list = ['financial_report_data','profit_data']
for i in range(0,len(list)):
    print("%s  %s " % (list[i],tab_list[i]))
    df = eval(list[i])(2017,3)
    print(df)
```
